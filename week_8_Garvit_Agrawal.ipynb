# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging
- Add more tools


````python
# 🛠️ TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
# 🤖 AGENT FUNCTION (IMPLEMENTED)

import re

def agent(query: str):
    try:
        if not query or not query.strip():
            return {
                "type": "error",
                "result": "Empty query received"
            }

        query_lower = query.lower()


        if "calculate" in query_lower:

            expr = re.sub(r"calculate", "", query, flags=re.IGNORECASE).strip()

            expr = re.sub(r"[^0-9+\-*/().\s]", "", expr).strip()

            if not expr:
                return {
                    "type": "error",
                    "result": "No valid expression found to calculate"
                }

            calc_result = calculator(expr)
            if calc_result == "Error in calculation":
                return {
                    "type": "error",
                    "result": calc_result
                }

            return {
                "type": "calculation",
                "result": calc_result
            }


        elif "keywords" in query_lower:

            text = re.sub(r"extract\s+keywords\s+from", "", query, flags=re.IGNORECASE)
            text = re.sub(r"keywords", "", text, flags=re.IGNORECASE).strip()

            if not text:
                return {
                    "type": "error",
                    "result": "No text found to extract keywords from"
                }

            keywords_result = extract_keywords(text)
            return {
                "type": "keywords",
                "result": keywords_result
            }


        else:
            return {
                "type": "general",
                "result": f"I received your query: '{query}'. This is a general response since no specific tool was matched."
            }

    except Exception as e:

        return {
            "type": "error",
            "result": f"Unexpected error: {str(e)}"
        }
````

## 📦 Expected Output Format

```
{
  "type": "calculation / keywords / general / error",
  "result": ...
}
```

````python
# 🧪 Test Cases

queries = [
    "Calculate 20 + 5",
    "Extract keywords from Artificial Intelligence is transforming industries",
    "What is machine learning?"
]

for q in queries:
    print("Query:", q)
    print("Response:", agent(q))
    print("-" * 50)
````
